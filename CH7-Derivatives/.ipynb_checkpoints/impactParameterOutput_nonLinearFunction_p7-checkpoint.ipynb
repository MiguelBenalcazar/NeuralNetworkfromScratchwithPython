```python
def f(x):
    return 2 * x ** 2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.array(range(5))
```

```python
y = f(x)
```

```python
print(x)
print(y)
```

```python
plt.title("Linear function y=2x^2 graphed")
plt.plot(x, y)
plt.show()
```
