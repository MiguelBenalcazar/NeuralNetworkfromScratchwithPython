```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# define function

def f(x):
    return 2 * x ** 2
```

```python
# np.arange(start, stop, step) to give us a smoother curve

x = np.array(np.arange(0, 5, 0.001))
y = f(x)
```

```python
plt.plot(x, y)
```

```python
colors = ['k', 'g', 'r', 'b', 'c']
```

```python
def approximate_tangent_line(x, approximate_derivative):
    return (approximate_derivative * x) + b
```

```python
for i in range(5):
    p2_delta = 0.0001 # deltaX
    bias = 0.9
    x1 = i
    x2 = x1 + p2_delta
    
    y1 = f(x1)
    y2 = f(x2)
    
    print(f'P1: ({x1}, {y1})  P2: ({x2}, {y2})')
    approximate_derivative = (y2 - y1)/(x2 - x1)
    b = y2 - (approximate_derivative * x2)
    to_plot = [x1 - bias, x1, x1 + bias]
    plt.scatter(x1, y1, c=colors[i])
    plt.plot([point for point in to_plot], [approximate_tangent_line(point, approximate_derivative) for point in to_plot], c=colors[i])
   
    print(f'Approximate derivative for f(x) where x={x1} is {approximate_derivative}')
    
plt.plot(x, y)  
plt.show()
```
